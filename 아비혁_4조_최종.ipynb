```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
#2019년의 월 별 모바일 이동량이 각 컬럼의 월 별 이동량 값의 중간값보다 큰 행만을 선별하여 데이터셋 만들기 
## '관광지' ~ '주거지등 칼럼 각 평균 : 9,715,838	9,411,779	12,559,151	199,987,625	97,749,205	161,852,233
###  1,4,5,6,8,10,11,12월이 대상으로 됨 
dataset = [['관광지', '대형아울렛', '레저스포츠', '사무지역','상업지역','주거지등'], ['관광지','레저스포츠','사무지역','주거지등'], ['관광지','대형아울렛','레저스포츠','사무지역','상업지역','주거지등'], ['관광지', '레저스포츠', '사무지역','주거지등'], ['관광지', '대형아울렛', '사무지역','상업지역','주거지등'], ['대형아울렛', '상업지역'], ['대형아울렛','상업지역']]
 
encoder = TransactionEncoder()
np_array = encoder.fit(dataset).transform(dataset)
print(np_array)
df = pd.DataFrame(np_array, columns = encoder.columns_)
display(df)
```

```python
frequent_itemsets = apriori(df, min_support = 0.5, use_colnames = True)
display(frequent_itemsets)
```

```python
association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1)
```
